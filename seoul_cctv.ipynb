```python
#인구대비 cctv 비율
data_result['cctv비율'] = data_result['소계']/data_result['인구수']*100
```

```python
data_result['cctv비율'].sort_values(ascending=False).plot(kind='bar',figsize=(12,3))
plt.show()
```

```python
#인구수와 cctv수 scatter plot
plt.figure(figsize=(12,3))
colormap = data_result['소계']
plt.scatter(data_result['인구수'],data_result['소계'],s=50,c=colormap,marker='o')
plt.colorbar()
plt.xlabel('인구수')
plt.ylabel('cctv')
plt.show()
```

```python
#인구수와 cctv에 대한 fitting curve
#1차식의 기울기와 y절편
fp1 = np.polyfit(data_result['인구수'],data_result['소계'],1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000,700000,100)
```

```python
#1차fitting curve 표시하기
#scatter: s=size, c=colormap, plt.colorbar()해야 함
#plot: ls=line spec, lw=line width
plt.figure(figsize=(12,3))
colormap = data_result['소계']
plt.scatter(data_result['인구수'],data_result['소계'],s=100,c=colormap,marker='o')
plt.colorbar()
plt.plot(fx,f1(fx),ls='dashed',lw=3,color='g')
plt.xlabel('인구수')
plt.ylabel('cctv')
plt.show()
```

```python
#추세선에서 오차가 크면 색이 달라지도록
#오차 계산 추가, 오차가 큰 순으로 정렬
data_result['오차'] = np.abs(data_result['소계'] - f1(data_result['인구수']))
df_sort = data_result.sort_values(by='오차',ascending=False)
df_sort.head()
```

```python
#scatter, fitting curve, 오차가 크면 색이 달라지도록, 점에 인덱싱
plt.figure(figsize=(12,3))
plt.scatter(data_result['인구수'],data_result['소계'],
           c=data_result['오차'],s=50)
plt.plot(fx,f1(fx),ls='dashed',lw=3,color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98,
            df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('cctv')
plt.colorbar()
plt.show()
```
